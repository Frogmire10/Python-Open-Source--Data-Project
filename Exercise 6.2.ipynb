```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load the cleaned dataset
file_path = '/Users/andrewfearney27/cleaned_battle_deaths.csv'  # Corrected path
data_cleaned = pd.read_csv(file_path)

#Display the first few rows of the dataset
print(data_cleaned.head())
```

```python
#Step 2: Select relevant variables for exploratory visual analysis
#Display first few rows of the dataset
print(data_cleaned.head())
```

```python
#Step 3: Create a correlation matrix heatmap
#Filter only numeric columns
numeric_columns = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data_cleaned[numeric_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()

#High Correlation between Battle Death Estimates: The variables bd_best, bd_low, and bd_high are highly correlated with each other, with correlation coefficients close to 1. 
#This is expected as these variables represent different estimates of battle deaths and are likely to move together.
```

```python
#Step 4: Create scatterplots for variables with the strongest correlations

#Scatterplot for bd_best and bd_low
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='bd_best', y='bd_low')
plt.title('Scatterplot of bd_best vs bd_low')
plt.show()

#Scatterplot for bd_best and bd_high
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='bd_best', y='bd_high')
plt.title('Scatterplot of bd_best vs bd_high')
plt.show()

#Scatterplot for bd_low and bd_high
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='bd_low', y='bd_high')
plt.title('Scatterplot of bd_low vs bd_high')
plt.show()

#bd_best vs bd_low: The scatterplot shows a strong linear relationship, reaffirming the high correlation seen in the heatmap. 
#This indicates that the bd_low estimate is very closely aligned with the bd_best estimate.
```

```python
#Step 5: Create a pair plot of the entire dataset
plt.figure(figsize=(12, 10))
sns.pairplot(data_cleaned)
plt.title('Pair Plot of the Entire Dataset')
plt.show()

#Diagonal Histograms: The histograms on the diagonal provide insights into the distribution of each variable. 
#For example, the year variable shows distinct clusters indicating specific periods of conflict data.
```

```python
#Step 6: Create a categorical plot for type_of_conflict vs bd_best
plt.figure(figsize=(10, 6))
sns.catplot(x='type_of_conflict', y='bd_best', data=data_cleaned, kind='box')
plt.title('Box Plot of type_of_conflict vs bd_best')
plt.show()

#Type of Conflict and Battle Deaths: The box plot indicates that different types of conflict (type_of_conflict) can have varying ranges of battle deaths. 
#For instance, some conflict types might consistently have higher or lower battle death estimates.
```
